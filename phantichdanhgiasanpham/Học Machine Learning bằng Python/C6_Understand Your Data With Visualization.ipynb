## 6.1. Univariate Plots

### 6.1.1. Histograms

```python
# Univariate Histograms
import matplotlib.pyplot as plt
import pandas as pd
# Tn file d liu
filename = 'pima-indians-diabetes-dataset.csv'
# i names cho ph hp vi file
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# c d liu
data = pd.read_csv(filename, names=names, header=0)
# V biu  histogram cho tng ct
data.hist(figsize=(10, 8), bins=20, edgecolor='black')
# Hin th biu 
plt.show()
```

## 6.2. Multivariate Plots

### 6.2.1. Correlation Matrix Plot

```python
# Correction Matrix Plot
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Tn file d liu
filename = 'pima-indians-diabetes-dataset.csv'
# i names cho ph hp vi file
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# c d liu
data = pd.read_csv(filename, names=names, header=0)
# Tnh ma trn tng quan
correlations = data.corr()
# V biu  ma trn tng quan
fig = plt.figure(figsize=(10, 8))  # iu chnh kch thc hnh
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1, cmap='coolwarm')
fig.colorbar(cax)
# Thit lp nhn
ticks = np.arange(0, len(names), 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names, rotation=45, fontsize=10, ha='left')  # Xoay v thu nh ch
ax.set_yticklabels(names, fontsize=10)
# Tng l di  trnh  nhn
plt.subplots_adjust(bottom=0.2, left=0.2)
plt.title("Correlation Matrix", pad=20)
plt.show()
```

```python
# Correction Matrix Plot (generic)
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Tn file d liu
filename = 'pima-indians-diabetes-dataset.csv'
# i names cho ph hp vi file
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# c d liu
data = pd.read_csv(filename, names=names, header=0)
correlations = data.corr()
# plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
plt.show()
```

### 6.2.2. Scatter Plot Matrix

```python
# Scatterplot Matrix
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import scatter_matrix  # Import ng
# Tn file d liu
filename = 'pima-indians-diabetes-dataset.csv'
# t tn ct ph hp
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# c d liu
data = pd.read_csv(filename, names=names, header=0)
# V scatter plot matrix
scatter_matrix(data, figsize=(12, 12), diagonal='hist')  # C th chn 'kde' thay 'hist'
plt.suptitle("Scatterplot Matrix", fontsize=16)
plt.show()
```
