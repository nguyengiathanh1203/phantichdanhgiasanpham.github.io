## 6.1. Univariate Plots

### 6.1.1. Histograms

```python
# Univariate Histograms
import matplotlib.pyplot as plt
import pandas as pd
# Tên file dữ liệu
filename = 'pima-indians-diabetes-dataset.csv'
# Đổi names cho phù hợp với file
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# Đọc dữ liệu
data = pd.read_csv(filename, names=names, header=0)
# Vẽ biểu đồ histogram cho từng cột
data.hist(figsize=(10, 8), bins=20, edgecolor='black')
# Hiển thị biểu đồ
plt.show()
```

### 6.1.2. Density Plots

```python
# Univariate Density Plots
import matplotlib.pyplot as plt
import pandas as pd
# Tên file dữ liệu
filename = 'pima-indians-diabetes-dataset.csv'
# Đổi names cho phù hợp với file
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# Đọc dữ liệu
data = pd.read_csv(filename, names=names, header=0)
# Vẽ biểu đồ mật độ (density) cho từng cột
data.plot(kind='density', subplots=True, layout=(3, 3), sharex=False, figsize=(10, 8))
# Hiển thị biểu đồ
plt.tight_layout()  # Tăng khoảng cách giữa các subplot để không bị chồng lấn
plt.show()
```

## 6.2. Multivariate Plots

### 6.2.1. Correlation Matrix Plot

```python
# Correction Matrix Plot
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Tên file dữ liệu
filename = 'pima-indians-diabetes-dataset.csv'
# Đổi names cho phù hợp với file
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# Đọc dữ liệu
data = pd.read_csv(filename, names=names, header=0)
# Tính ma trận tương quan
correlations = data.corr()
# Vẽ biểu đồ ma trận tương quan
fig = plt.figure(figsize=(10, 8))  # Điều chỉnh kích thước hình
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1, cmap='coolwarm')
fig.colorbar(cax)
# Thiết lập nhãn
ticks = np.arange(0, len(names), 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names, rotation=45, fontsize=10, ha='left')  # Xoay và thu nhỏ chữ
ax.set_yticklabels(names, fontsize=10)
# Tăng lề dưới để tránh đè nhãn
plt.subplots_adjust(bottom=0.2, left=0.2)
plt.title("Correlation Matrix", pad=20)
plt.show()
```

```python
# Correction Matrix Plot (generic)
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Tên file dữ liệu
filename = 'pima-indians-diabetes-dataset.csv'
# Đổi names cho phù hợp với file
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# Đọc dữ liệu
data = pd.read_csv(filename, names=names, header=0)
correlations = data.corr()
# plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
plt.show()
```

### 6.2.2. Scatter Plot Matrix

```python
# Scatterplot Matrix
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import scatter_matrix  # Import đúng
# Tên file dữ liệu
filename = 'pima-indians-diabetes-dataset.csv'
# Đặt tên cột phù hợp
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# Đọc dữ liệu
data = pd.read_csv(filename, names=names, header=0)
# Vẽ scatter plot matrix
scatter_matrix(data, figsize=(12, 12), diagonal='hist')  # Có thể chọn 'kde' thay 'hist'
plt.suptitle("Scatterplot Matrix", fontsize=16)
plt.show()
```
