## Rescale Data

```python
# Rescale data (between 0 and 1)
import pandas as pd
from numpy import set_printoptions
from sklearn.preprocessing import MinMaxScaler
# Tên file dữ liệu
filename = 'pima-indians-diabetes-dataset.csv'
# Đặt tên cột phù hợp
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# Đọc dữ liệu
data = pd.read_csv(filename, names=names, header=0)
# Tách dữ liệu thành đầu vào và đầu ra
array = data.values
X = array[:, 0:8]
Y = array[:, 8]
# Chuẩn hóa dữ liệu về khoảng [0, 1]
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX = scaler.fit_transform(X)
# Hiển thị dữ liệu đã chuẩn hóa
set_printoptions(precision=3)
print(rescaledX[0:5, :])
```

## Normalize Data

```python
from sklearn.preprocessing import Binarizer
from pandas import read_csv
from numpy import set_printoptions
# Tên file dữ liệu
filename = 'pima-indians-diabetes-dataset.csv'  # Giữ nguyên tên file
# Định nghĩa tên cột theo đúng cấu trúc của file CSV
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# Đọc dữ liệu
data = pd.read_csv(filename, names=names, header=0)
array = data.values
# separate array into input and output components
X = array[:,0:8]
Y = array[:,8]
binarizer = Binarizer(threshold=0.0).fit(X)
binaryX = binarizer.transform(X)
# summarize transformed data
set_printoptions(precision=3)
print(binaryX[0:5,:])
```
