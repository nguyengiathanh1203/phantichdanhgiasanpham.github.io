# Feature Selection For Machine Learning

## Univariate Selection

```python
# Univariate Density Plots
import matplotlib.pyplot as plt
import pandas as pd
# Tn file d liu
filename = 'pima-indians-diabetes-dataset.csv'
# t tn ct ph hp vi cu trc ca file CSV
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# c d liu, b dng tiu  nu  c sn trong file
data = pd.read_csv(filename, names=names, header=0)
# V biu  mt  (density) cho tng ct
data.plot(kind='density', subplots=True, layout=(3, 3), sharex=False, figsize=(10, 8))
# Hin th biu 
plt.tight_layout()  # Tng khong cch gia cc subplot  khng b chng ln
plt.show()
```

## Recursive Feature Elimination

```python
# Feature Extraction with RFE
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
# Tn file d liu
filename = 'pima-indians-diabetes-dataset.csv'
# t tn ct ph hp
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# c d liu
dataframe = pd.read_csv(filename, names=names, header=0)
array = dataframe.values
X = array[:, 0:8]
Y = array[:, 8]
# Feature extraction
model = LogisticRegression(max_iter=1000)
rfe = RFE(model, n_features_to_select=3)
fit = rfe.fit(X, Y)
# Hin th kt qu
print("Num Features: %d" % fit.n_features_)
print("Selected Features: %s" % fit.support_)
print("Feature Ranking: %s" % fit.ranking_)
```

## Principal Component Analysis

```python
# Feature Extraction with PCA
import pandas as pd
from sklearn.decomposition import PCA
# Tn file d liu
filename = 'pima-indians-diabetes-dataset.csv'
# t tn ct ph hp
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# c d liu
dataframe = pd.read_csv(filename, names=names, header=0)
array = dataframe.values
X = array[:, 0:8]
Y = array[:, 8]
# Feature extraction
pca = PCA(n_components=3)
fit = pca.fit(X)
# Hin th kt qu
print("Explained Variance: %s" % fit.explained_variance_ratio_)
print("Principal Components (loadings):\n", fit.components_)
```

## Feature Importance

```python
# Feature Importance with Extra Trees Classifier
import pandas as pd
from sklearn.ensemble import ExtraTreesClassifier
# Tn file d liu
filename = 'pima-indians-diabetes-dataset.csv'
# t tn ct ph hp
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# c d liu
dataframe = pd.read_csv(filename, names=names, header=0)
array = dataframe.values
X = array[:,0:8]
Y = array[:,8]
# feature extraction
model = ExtraTreesClassifier()
model.fit(X, Y)
print(model.feature_importances_)
```
