# Feature Selection For Machine Learning

## Univariate Selection

```python
# Univariate Density Plots
import matplotlib.pyplot as plt
import pandas as pd
# Tên file dữ liệu
filename = 'pima-indians-diabetes-dataset.csv'
# Đặt tên cột phù hợp với cấu trúc của file CSV
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# Đọc dữ liệu, bỏ dòng tiêu đề nếu đã có sẵn trong file
data = pd.read_csv(filename, names=names, header=0)
# Vẽ biểu đồ mật độ (density) cho từng cột
data.plot(kind='density', subplots=True, layout=(3, 3), sharex=False, figsize=(10, 8))
# Hiển thị biểu đồ
plt.tight_layout()  # Tăng khoảng cách giữa các subplot để không bị chồng lấn
plt.show()
```

## Recursive Feature Elimination

```python
# Feature Extraction with RFE
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
# Tên file dữ liệu
filename = 'pima-indians-diabetes-dataset.csv'
# Đặt tên cột phù hợp
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# Đọc dữ liệu
dataframe = pd.read_csv(filename, names=names, header=0)
array = dataframe.values
X = array[:, 0:8]
Y = array[:, 8]
# Feature extraction
model = LogisticRegression(max_iter=1000)
rfe = RFE(model, n_features_to_select=3)
fit = rfe.fit(X, Y)
# Hiển thị kết quả
print("Num Features: %d" % fit.n_features_)
print("Selected Features: %s" % fit.support_)
print("Feature Ranking: %s" % fit.ranking_)
```

## Principal Component Analysis

```python
# Feature Extraction with PCA
import pandas as pd
from sklearn.decomposition import PCA
# Tên file dữ liệu
filename = 'pima-indians-diabetes-dataset.csv'
# Đặt tên cột phù hợp
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# Đọc dữ liệu
dataframe = pd.read_csv(filename, names=names, header=0)
array = dataframe.values
X = array[:, 0:8]
Y = array[:, 8]
# Feature extraction
pca = PCA(n_components=3)
fit = pca.fit(X)
# Hiển thị kết quả
print("Explained Variance: %s" % fit.explained_variance_ratio_)
print("Principal Components (loadings):\n", fit.components_)
```

## Feature Importance

```python
# Feature Importance with Extra Trees Classifier
import pandas as pd
from sklearn.ensemble import ExtraTreesClassifier
# Tên file dữ liệu
filename = 'pima-indians-diabetes-dataset.csv'
# Đặt tên cột phù hợp
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
# Đọc dữ liệu
dataframe = pd.read_csv(filename, names=names, header=0)
array = dataframe.values
X = array[:,0:8]
Y = array[:,8]
# feature extraction
model = ExtraTreesClassifier()
model.fit(X, Y)
print(model.feature_importances_)
```
